```python
import pandas as pd
import numpy as np
import datetime
import sklearn as sklearn
from sklearn.impute import SimpleImputer
import dateparser
```

Step-1: Read the csv file

```python
df= pd.read_csv("data4preprocessing.csv")
```

Step-2: Process the na/missing entries of the second column using mean imputation

```python
cols=list(df.columns)
col1=cols[0]
col2=cols[1]
vec=df[col2]
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
imp.fit(np.array(vec).reshape(1,-1))
arr=imp.transform(np.array(vec).reshape(1,-1))
arr2=np.transpose(arr)
df[col2]=pd.Series(np.ndarray.flatten(arr2))
```

```python
val=pd.Series([])
indx=pd.Series([])
for i in list(df.index.values):
    val[i]=dateparser.parse(df.loc[i,cols[0]])
    if pd.isnull(val[i]):
        indx=indx.append(pd.Series([i]))
pd.Series(indx.values)
df[cols[0]]=val
```

```python
df=df.drop(indx,axis=0)
df=df.groupby(cols[0], group_keys=False).mean().reset_index()
```

```python
df=df.sort_values(cols[0])
mean_val=df[col2].mean()
mean_val
```

Finding the frequency of the timeseries data as

```python
frequency=df[cols[0]].diff().value_counts().idxmax()
```

```python
df2=pd.date_range(start=df[cols[0]].min(), end=df[cols[0]].max(),freq=frequency)
```

```python
val=[mean_val]*len(df2)
val
d={cols[0]:df2,cols[1]:val}
final_data=pd.DataFrame(data=d)
final_data
```

```python
import pandas as pd
import datetime
import dateparser
```

```python
df= pd.read_csv("final_data.csv")
cols=list(df.columns)
df=pd.DataFrame.drop(df,labels=cols[0],axis=1)
```

```python
df[cols[0]]=pd.to_datetime(df[cols[0]], infer_datetime_format=True)
df
```

```python
cols=list(df.columns)
cols
```

```python
def hours(x):
    return x.hour
def minutes(x):
    return x.minute
def seconds(x):
    return x.second
def years(x):
    return x.year
def months(x):
    return x.month
def days(x):
    return x.day
```

```python
df['Hour']=df[cols[0]].apply(hours)
df['Minute']=df[cols[0]].apply(minutes)
df['Second']=df[cols[0]].apply(seconds)
df['Year']=df[cols[0]].apply(years)
df['Month']=df[cols[0]].apply(months)
df['Day']=df[cols[0]].apply(days)
df
```

```python
df.to_csv("time_components_collected.csv")
```
